## Import packages

```python
import numpy as np
import pandas as pd

import sklearn as sk
from sklearn.cross_validation import train_test_split
from sklearn.grid_search import GridSearchCV
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

import xgboost as xgb
from scipy import interp

import random
from __future__ import print_function

from jllib import grid_search, xval, feat_imp, viz

import warnings
warnings.filterwarnings('ignore')
```

## Define parameters

```python
INPUT_PATH = 'C:\Users\liuleo\Documents\KT\Python_template\Classification\exa_dataset.csv' #"path saving your model building dataset"
FILE_SEP = '|' #"input file separator, pipe for example"
TARGET = 'm_target' #'name of the column for classification target'
ID = 'custid'#'row id (for example customer id)'
PRED_PROBA = 'm_target_proba'#'column name to store preidction score'
```

## Read files

```python
df = pd.read_csv(INPUT_PATH, sep=FILE_SEP, na_values=["\N", "NULL"]) 
```

## Get feature columns

```python
feature_id_columns = [x for x in df.columns if x != TARGET]
feature_columns = [x for x in df.columns if (x not in [TARGET, ID])]
print("number of features to use {}".format(len(feature_columns)))
```

## Check target percentage in population

```python
print(TARGET)
print("target number %s"%TARGET)
print(df[TARGET].sum())
print("target proportion %s"%TARGET)
print(df[TARGET].mean())
```

## Split dataset for Grid Search

```python
train, test = train_test_split(df,train_size=0.7)

print('Train data has %i rows and %i columns'%(train.shape[0], train.shape[1]))
print('Test data has %i rows and %i columns'%(test.shape[0], test.shape[1]))
```

## RF Grid Search

### Define hyper-parameters combinations to search

```python
grid_parameters = {'n_estimators':[200],'min_samples_split':[5], 'min_samples_leaf':[5]} 
#criterion='gini',max_depth=None, max_features='auto',class_weight=None, 
clf_rf_base=RandomForestClassifier(n_jobs=4,random_state=27,verbose=0)
clf = GridSearchCV(clf_rf_base,grid_parameters,scoring='roc_auc',cv=3,verbose=1,n_jobs=4)
```

### Call Grid Search fucntion to find the best fit parameters 

```python
clf_gs, results_test = grid_search.grid_search_trees(clf, train, test, ID, TARGET, PRED_PROBA)
```

### Feature importance for best estimator

```python
feature_importances_sort = feat_imp.feat_imp_trees(clf_gs.best_estimator_, feature_columns)
```

### Cross-validation on entire dataset with best estimator

```python
#Random Forest
classifier = RandomForestClassifier(n_estimators=200,n_jobs=4, random_state=27,min_samples_split=5, min_samples_leaf=5, 
                                    verbose=0)
results_cv_targeting, feature_importances, nb_calls_cv = xval.run_cross_validation_rf(df, classifier , feature_id_columns,ID,
                                                                                    TARGET,PRED_PROBA,5,.3,27)
```

### Feature importance for cross-validation

```python
feat_imp_sort_df = feat_imp.feat_imp_xval_plot(feature_importances)
```

## Xgboost Grid Search

### Define hyper-parameters combinations to search

```python
grid_parameters = {'max_depth': [6],'n_estimators': [170],'learning_rate':[0.05],'min_child_weight':[25],'max_delta_step':[1],
                   'gamma':[0.1],'scale_pos_weight':[1], 'colsample_bytree':[0.85],'subsample':[0.85],'colsample_bylevel':[0.85]}

clf_xgb_base= xgb.XGBClassifier(nthread=2, seed=27)
clf = GridSearchCV(clf_xgb_base,grid_parameters,scoring='roc_auc',cv=3,verbose=1,n_jobs=4) 
```

### Call Grid Search fucntion to find the best fit parameters

```python
clf_gs, results_test = grid_search.grid_search_trees(clf, train, test, ID, TARGET, PRED_PROBA)
```

### Feature importance for best estimator

```python
feature_importances_sort = feat_imp.feat_imp_xgb(clf_gs.best_estimator_, feature_columns)
```

### Cross-validation on entire dataset with best estimator

```python
classifier = xgb.XGBClassifier(objective='binary:logistic',max_depth=6,n_estimators=170, learning_rate=0.05,max_delta_step=1,
                        min_child_weight=25, gamma=0.1, scale_pos_weight=1, colsample_bytree=0.85, subsample=0.85,colsample_bylevel=0.85,
                        nthread=6, seed=27, silent=1)

results_cv_targeting, feature_importances,nb_calls_cv = xval.run_cross_validation_xgb(df, classifier, feature_id_columns, ID, 
                                                                                      TARGET,PRED_PROBA,5,.3,27, False, 40)
```

### Feature importance for cross-validation

```python
feat_imp_sort_df = feat_imp.feat_imp_xval_plot(feature_importances)
```

## Logistic regression Grid Search

### Define hyper-parameters combinations to search

```python
grid_parameters = {'penalty':['l2'], 'C':[0.5], 'class_weight':[None]}
clf_lg_base=LogisticRegression(n_jobs=1,random_state=27,verbose=0)
clf = GridSearchCV(clf_lg_base,grid_parameters,scoring='roc_auc',cv=3,verbose=1,n_jobs=1)
```

### Call Grid Search fucntion to find the best fit parameters

```python
clf_gs, results_test = grid_search.grid_search_linear(clf, train, test, ID, TARGET, PRED_PROBA, True)
```

### Feature importance for best estimator

```python
feature_importances_sort = feat_imp.feat_imp_linear(clf_gs.best_estimator_, feature_columns)
```

### Cross-validation on entire dataset with best estimator

```python
classifier = LogisticRegression(penalty='l2', C=0.5, class_weight=None, random_state=27)

results_cv_targeting, feature_importances, nb_calls_cv = xval.run_cross_validation_linear(df, classifier , feature_id_columns,ID,
                                                                                     TARGET,PRED_PROBA,5,.3,27, True)
```

### Feature importance for cross-validation

```python
feat_imp_sort_df = feat_imp.feat_imp_xval_plot(feature_importances)
```
