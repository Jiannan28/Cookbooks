## Handle categorical features for models

```python
### CONVERT CATEGORICAL VARIABLE TO DUMMIES ###
cat_variables = ['Age_group', 'Sex', 'Embarked']
dummies = pd.get_dummies(train[cat_variables])
dummy_cols = dummies.columns

### REPLACE VARIABLE BY ITS DUMMIES ###
train_pr = train.join(dummies)
train_pr.drop(cat_variables, axis=1, inplace=True)
train_pr.drop('Sex_male', axis=1, inplace=True)
print train_pr.columns
#print train_pr.ix[train.Embarked.isnull(), ['Embarked_C', 'Embarked_Q', 'Embarked_S']]
```

## Train - test

```python
### CHECK SAMPLE SIZE ###
dim =len(train_pr)
print dim
```

```python
from sklearn.cross_validation import train_test_split

### DEFINE SIZE OF TEST SAMPLE ###
train_size = 0.5

### SPLIT TRAIN AND TEST ###
tr = train_pr.sample(frac=train_size, random_state=120798)
te = train_pr.drop(tr.index, 0)

### SEPARATE COVARIATES AND TARGET VARIABLE ###
X_tr, y_tr = tr.drop(target, 1), tr[target]
X_te, y_te = te.drop(target, 1), te[target]
print 'train sample size :', len(X_tr)
print 'test sample size :', len(X_te)
```

If the size of the dataset is large enough, do not hesitate to reduce the size of the training set and to increase the size of the test set. For example a 20/80 is sometimes the better option at the begining of your modeling journey !

## Decision tree

```python
### SELECT FEATURES ###
modelVars = [ x for x in ["Pclass", "SibSp", "Parch"]+list(dummy_cols) if x != 'Sex_male']
print modelVars
```

```python
from sklearn import tree

### TRAIN THE TREE MODEL ###
clf_tree = tree.DecisionTreeClassifier(max_depth=5)

clf_tree.fit(X_tr[modelVars], y_tr)

params_clf = {"min_samples_leaf":5}
clf_tree.set_params(**params_clf)
```

Visualize decision tree with d3js

```python
def rules(clf, features, labels, node_index=0):
    """Structure of rules in a fit decision tree classifier

    Parameters
    ----------
    clf : DecisionTreeClassifier
        A tree that has already been fit.

    features, labels : lists of str
        The names of the features and labels, respectively.

    """
    node = {}
    if clf.tree_.children_left[node_index] == -1:  # indicates leaf
        count_labels = zip(clf.tree_.value[node_index, 0], labels)
        node['name'] = ', '.join(('{} of {}'.format(int(count), label)
                                  for count, label in count_labels))
    else:
        feature = features[clf.tree_.feature[node_index]]
        threshold = clf.tree_.threshold[node_index]
        node['name'] = '{} > {}'.format(feature, threshold)
        left_index = clf.tree_.children_left[node_index]
        right_index = clf.tree_.children_right[node_index]
        node['children'] = [rules(clf, features, labels, right_index),
                            rules(clf, features, labels, left_index)]
    return node

import json
r = rules(clf_tree, modelVars, 'Survived')
with open('C://Users/liuleo/Documents/KT/Python_template/LightTraining/rules.json', 'w') as f:
    f.write(json.dumps(r))
```

## Prediction

### COMPUTE PREDICTED PROBABILITIES ###

```python
pred = clf_tree.predict_proba(X_te[modelVars])
pred = pd.Series(pred[:,1], index=X_te.index, name='pred_proba')
test_tree = X_te[modelVars].join(pred).join(y_te)

test_tree.sample(10)
```

## Measure the performance

### Accuracy of prediction

```python
from sklearn.metrics import *

### CHOOSE THRESHOLD FOR PREDICTION ###
test_tree['my_prediction'] = (test_tree['pred_proba'] > 0.38)+0

### COMPUTE CONFUSION MATRIX ###
conf_matrix = confusion_matrix(test_tree['Survived'], test_tree['my_prediction'])

print pd.DataFrame(conf_matrix, index=['predicted_died', 'predicted_survived'], columns=['died', 'survived'])
```

```python
### COMPUTE PREDICTION ACCURACY ###

# Accuracy = (True Positive + True Negative)/Total Population
accuracy = sum(conf_matrix.diagonal())/float(sum(conf_matrix))
print accuracy
```

<font color='blue'> It's always difficult to find the right **threshold**. It should be a business driven decision !

## ROC curve

<font color='blue'> The **ROC curve** allows the evaluation of a model without specifying the threshold

```python
from scipy import interp

### COMPUTE ROC CURVE POINTS ###
fpr, tpr, thres = roc_curve(test_tree['Survived'], test_tree['pred_proba'])

### INTERPOLATE VALUES ###
fpr_grid = np.linspace(0, 1, 101)
completed_tpr = interp(fpr_grid, fpr, tpr)
auc_score = auc(fpr, tpr)

### ROC CURVE OF RANDOM CLASSIFIER VS OUR CLASSIFIER ###
plt.figure(figsize=(15,10))
plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Random')
plt.plot(fpr_grid, completed_tpr, 'k--',
         label='ROC (area = %0.2f)' % auc_score, lw=2)
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic for Tree Classifier',fontsize=17)
plt.legend(loc="lower right")
plt.show()
```

# ROC curve

## Check some points of the curve

```python
### CHECK ROC CURVE FOR SOME THRESHOLD VALUE ###
thres_values = [0.9, 0.3]
for t in thres_values:
    idx = min(range(len(thres)), key=lambda i: abs(thres[i]-t))
    fpr_value, tpr_value = fpr[idx], tpr[idx]
    print 'for threshold ', t, ': fpr =', fpr_value, ', tpr =', tpr_value
```

## Area under the curve

Because curves are not always easy to compare, we prefer the Area under the ROC curve instead of the ROC curve itself.

# Random Forest / Gradient Boosting Tree

### The main idea behind a **Random Forest**

- Multiple trees each time with a random sample (with replacement) of the observations (bagging/bootstrap)
- For each tree and for each node the variables are choseen randomly
- At the end the prediction is the majority of the predicted value for each tree
- It's horizontal multiple trees concept

### The main idea behind a **Gradient Boosting**

- Multiple trees one after the other
- Each new tree is train on the residuals of all the previous
- The learning power of each tree is penalized : multiple weak learner
- Variables and obervation can be random at some stage (like Random Forest)
- It's vertical multiple trees concept

# How to increase a model performance

**2 steps**

- **Feature engineering** (very important):
    - Adding new variables (internal and external)
    - Create new variables with existing one :
        - Put Past information to the present
        - Create rich expert KPI
    - Select appropriate features (features selection) :
        - Use statistics like correlation o CramerV (https://stackoverflow.com/questions/20892799/using-pandas-calculate-cram%C3%A9rs-coefficient-matrix)
        - Use data vizualisation
        - Use directly the learning algorithm (regularization)
        
        
- **Fine Tune** the model parameters (less important):
    - Train with different parameters and evaluate on test dataset
    - Repeat this many times on different (train/test) sampling (Cross-validation)

# Feature engineering

## Family, fare and cabin

```python
### sibsp OR parch IS family ###
train_pr['Family_size'] = train_pr['SibSp']+train_pr['Parch']+1
train_pr['Fare_group'] = pd.cut(train_pr['Fare'], bins=[0, 10, 20, 30, float('inf')], right=False, labels=[1, 2, 3, 4])
```

## Cabin category

```python
### EXTRACT FIRST LETTER OF THE CABIN ###
filled_values = train_pr['Cabin'].notnull()
train_pr['Cabin_cat'] = train_pr['Cabin']
train_pr.ix[filled_values, 'Cabin_cat'] = train_pr['Cabin'][filled_values].apply(lambda x: x[0:1])
```

## Title

```python
### EXTRACT TITLE FROM THE NAME ###
train_pr['Title'] = train_pr['Name'].apply(lambda x: x.split(', ')[1].split('.')[0])

### GROUP SOME TITLES ###
idx1 = train_pr['Title'].apply(lambda x: x in ['Capt', 'Don', 'Major', 'Sir'])
train_pr.ix[idx1, 'Title'] = 'Sir'
idx2 = train_pr['Title'].apply(lambda x: x in ['Dona', 'Lady', 'the Countess', 'Jonkheer'])
train_pr.ix[idx2, 'Title'] = 'Lady'
idx3 = train_pr['Title'].apply(lambda x: x in ['Miss', 'Mlle'])
train_pr.ix[idx3, 'Title'] = 'Miss'
idx4 = train_pr['Title'].apply(lambda x: x in ["Mrs", 'Ms', "Mme"])
train_pr.ix[idx4, 'Title'] = 'Mrs'
idx5 = train_pr['Title'].apply(lambda x: x in ['Col', 'Dr'])
train_pr.ix[idx5, 'Title'] = 'Other'
```

## Visualization

```python
f, axes = plt.subplots(1, 2, figsize = (16,6))
sns.barplot(x="Family_size", y="Survived", data=train_pr, ax=axes[0], ci=None) # not specify confidence interval (default based on std dev)
axes[0].set_title("Survival chances vs family size", fontsize=17)
sns.barplot(x="Title", y="Survived", data=train_pr, ax=axes[1], ci=None)
axes[1].set_title("Survival chances vs title", fontsize=17)
plt.show()
```

## New Model

```python
### CONVERT CATEGORICAL VARIABLE TO DUMMIES ###
cat_variables = ['Cabin_cat', 'Title']
dummies = pd.get_dummies(train_pr[cat_variables])
dummy_cols = dummies.columns

### REPLACE VARIABLE BY ITS DUMMIES ###
train_pr2 = train_pr.join(dummies)
train_pr2.drop(cat_variables, 1, inplace=True)

### SPLIT TRAIN AND TEST ###
tr = train_pr2.sample(frac=train_size, random_state=120798)
te = train_pr2.drop(tr.index, 0)

### SEPARATE COVARIATES AND TARGET VARIABLE ###
X_tr, y_tr = tr.drop(target, 1), tr[target]
X_te, y_te = te.drop(target, 1), te[target]

### TRAIN GRADIENT BOOSTING MODEL ###
modelVars = ["Pclass", "SibSp", "Parch", "Family_size", "Fare_group"]+list(dummy_cols)
gbc2 = GradientBoostingClassifier(n_estimators=75, learning_rate=0.03, max_depth=5)
gbc2.fit(X_tr[modelVars], y_tr)
```

## Performance

```python
pred = gbc2.predict_proba(X_te[modelVars])
pred = pd.Series(pred[:,1], index=X_te.index, name='pred_proba')
test_gbc2 = X_te[modelVars].join(pred).join(y_te)
```

```python
print roc_auc_score(test_gbc2['Survived'], test_gbc2['pred_proba'])
```
